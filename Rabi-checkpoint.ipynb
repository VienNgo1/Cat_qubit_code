```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt

import sim_utils2 as su
MHz = su.MHz
```

```python
cq = su.CQ(nqubits=1)

cq.add_rabi_oscillation(0)

dt = 0.001 / MHz
t = np.arange(0., 0.5/MHz, dt)

result = qt.mesolve(cq.H, cq.cat_states()[0], t,
                    c_ops=cq.loss_terms(),
                    e_ops=[psi * psi.dag() for psi in cq.cat_states()],
                    options=qt.Options(store_states=False, store_final_state=True))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(t * MHz, result.expect[1], color="black")
plt.plot(t * MHz, result.expect[0], color="red")

plt.xlabel("t [$\mu$s]", fontsize=14)
plt.ylabel("Tr {$|C_\\alpha^+ \\rangle \langle C_\\alpha^+ | \\rho$}", fontsize=14)
plt.tight_layout()
plt.savefig("rabi.pdf")
```
