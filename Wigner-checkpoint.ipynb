```python
import numpy as np
import matplotlib.pyplot as plt
import qutip as qt

import sim_utils2 as su
MHz = su.MHz
```

```python
cq = su.CQ(nqubits=1)

dt = 0.001 / MHz
t = np.arange(0., 0.5/MHz, dt)

result = qt.mesolve(cq.H, cq.cat_states()[0], t,
                    c_ops=cq.loss_terms(),
                    e_ops=[psi * psi.dag() for psi in cq.cat_states()],
                    options=qt.Options(store_states=False, store_final_state=True))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(t * MHz, result.expect[0], color="black")#, label="$|\pm X \\rangle$")
# plt.plot(t * MHz, 0.5 * np.tanh(4. * t * MHz / cq.tau - 2.) + 0.5, "--", color="red")
plt.plot(t * MHz, result.expect[1], color="red")
plt.xlabel("t [$\mu$s]", fontsize=14)
plt.ylabel("Tr {$|C_\\alpha^+ \\rangle \langle C_\\alpha^+ | \\rho$}", fontsize=14)
plt.tight_layout()
#plt.legend(fontsize=14)
plt.savefig("initialization.pdf")
```

```python
x = np.linspace(-4., 4., 100)

plt.figure()
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
w = qt.wigner(cq.thermal_state(), x[1:], x[1:])
w /= np.max(np.abs(w))
fig = plt.pcolormesh(x, x, w, cmap="bwr", vmin=-1., vmax=1.)
#plt.colorbar(fig)
plt.tight_layout()
plt.savefig("ground.pdf")

plt.figure()
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
w = qt.wigner(result.final_state, x[1:], x[1:])
w /= np.max(np.abs(w))
fig = plt.pcolormesh(x, x, w, cmap="bwr", vmin=-1., vmax=1.)
#plt.colorbar(fig)
plt.tight_layout()
plt.savefig("initialized_cat.pdf")

plt.figure()
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
w = qt.wigner(cq.cat_states()[0], x[1:], x[1:])
w /= np.max(np.abs(w))
fig = plt.pcolormesh(x, x, w, cmap="bwr", vmin=-1., vmax=1.)
#plt.colorbar(fig)
plt.tight_layout()
plt.savefig("ideal_cat.pdf")
```
